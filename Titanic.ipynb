#Filtering

```python
RMS
```

```python
class_filter = RMS["class"] == "3rd"
class_filter
```

```python
RMS_pass = RMS[(RMS["class"] == "3rd") | (RMS["class"] == "2nd") | (RMS["class"] == "1st")]
RMS_pass
```

##Plotting

```python
female_filter = RMS["gender"] == "female"
female_filter
```

```python
female_filter_age = RMS[female_filter]["age"]
female_filter_age
```

```python
sns.displot(data=RMS_pass, x="age", col="gender")
```

```python
sns.displot(data=RMS_pass, x="age", hue="gender", kind="kde")
```
